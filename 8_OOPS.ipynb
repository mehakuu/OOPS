```python
class Number:
    def sum(self):
        return self.a + self.b
```

```python
num = Number() 
num.a = 12
num.b = 34
s = num.sum()  
print(s)
```

```python
class railwayform:
    def printdata(self):
        print(self.name)
        print(self.train)
```

```python
pardeep= railwayform()
pardeep.name= "Pardeep Kumar"
pardeep.train= "Rajdhani"
pardeep.printdata()  
```

```python
Akeel = railwayform()
Akeel.name ="Akeel farooq"
Akeel.train= "Shatabdi Express"
Akeel.printdata()
```

```python
#class attribute
class employee:
    Company="Google"
    Salary=100
```

```python
venkata =employee()
chirag = employee()
print(venkata.Company)
print(venkata.Salary)
print(chirag.Company)
print(chirag.Salary)
```

```python
employee.Company="Facebook"
print(venkata.Company)
print(venkata.Salary)
print(chirag.Company)
print(chirag.Salary)
```

```python
#instance attribute 
venkata.Salary=500
chirag.Salary=700
print(venkata.Company)
print(venkata.Salary)
print(chirag.Company)
print(chirag.Salary)
```
